## Étape 2 : Vérifier les Fichiers


```python
import os

required_files = [
    'app.py',
    'requirements.txt',
    '2025-08-15_composition_sp500.csv',
    '2025-09-26_stocks-performance.csv'
]

print("📋 Vérification des fichiers nécessaires :")

all_present = True
for file in required_files:
    if os.path.exists(file):
        print(f"✅ {file}")
    else:
        print(f"❌ {file} - MANQUANT")
        all_present = False

print("=" * 50)

if all_present:
    print("\n✅ Tous les fichiers sont présents !")
else:
    print("\n⚠️ Certains fichiers manquent.")
```

## Étape 3 : Lancer l'Application Streamlit


```python
import subprocess
import sys
import os
import time

# Vérifier que app.py existe
if not os.path.exists('app.py'):
    print("❌ Erreur: app.py non trouvé")
    print(f"Répertoire actuel: {os.getcwd()}")
else:
    print("🚀 Lancement de Streamlit...")
    
    # Lancer Streamlit
    process = subprocess.Popen(
        [sys.executable, "-m", "streamlit", "run", "app.py"],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        universal_newlines=True
    )
    
    time.sleep(2)
    
    print("\n✅ Streamlit lancé !")
    print("\n🌐 Ouvrez votre navigateur et allez sur :")
    print("   👉 http://localhost:8501")
    
    # Afficher un lien cliquable si dans un notebook Jupyter
    try:
        from IPython.display import HTML, display
        display(HTML('<p><a href="http://localhost:8501" target="_blank" style="font-size: 18px; color: #10b981;">🎯 Ouvrir ReguAI → http://localhost:8501</a></p>'))
    except:
        pass
    
    print(f"\nProcess ID: {process.pid}")
    print("\n💡 Pour arrêter l'app, fermez cette cellule ou redémarrez le kernel.")
```

## ✅ Application Lancée !

L'application devrait maintenant être accessible sur **http://localhost:8501**
